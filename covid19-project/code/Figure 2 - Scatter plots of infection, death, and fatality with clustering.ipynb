```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### Figure 2

#### Figure 2A

```
sydata = pd.read_csv("final file/data/sycluster.csv")
sydata['Positives Case per 100K Population'] = sydata['Positives Case per 10K Population'] * 10
sydata['Deaths per 100K Population'] = sydata['Deaths per 10K Population'] * 10
sydata['County'] = sydata['County'].replace('New York', 'Manhattan')

plt.figure(figsize=(12,10))
plt.scatter(y = sydata['Deaths per 100K Population'], x = sydata['Positives Case per 100K Population'], s=50, alpha=0.5)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.ylabel('Death (per 100K Population)', fontsize=20)
plt.xlabel('Infection (per 100K Population)', fontsize=20)
plt.show()
```

#### Figure 2B

```
sydata = pd.read_csv("/covid-19-project/final file/data/sycluster.csv")
sydata['Positives Case per 100K Population'] = sydata['Positives Case per 10K Population'] * 10
sydata['Deaths per 100K Population'] = sydata['Deaths per 10K Population'] * 10
sydata['County'] = sydata['County'].replace('New York', 'Manhattan')

kmeans = KMeans(n_clusters=3, random_state = 42).fit(sydata[['Positives Case per 100K Population','Deaths per 10K Positive Case']])

centroids = kmeans.cluster_centers_

sydata['PosDeathCluster'] = kmeans.predict(sydata[['Positives Case per 100K Population','Deaths per 10K Positive Case']])
sydata['PosDeathCluster'] = sydata['PosDeathCluster'].replace({0: 2, 2: 1, 1: 3})

plt.figure(figsize=(12,10))
plt.scatter(y = sydata[sydata['PosDeathCluster'] == 1]['Deaths per 10K Positive Case'], x = sydata[sydata['PosDeathCluster'] == 1]['Positives Case per 100K Population'], s=50, alpha=0.7, color = 'red')
plt.scatter(y = sydata[sydata['PosDeathCluster'] == 2]['Deaths per 10K Positive Case'], x = sydata[sydata['PosDeathCluster'] == 2]['Positives Case per 100K Population'], s=50, alpha=0.7, color = 'orange')
plt.scatter(y = sydata[sydata['PosDeathCluster'] == 3]['Deaths per 10K Positive Case'], x = sydata[sydata['PosDeathCluster'] == 3]['Positives Case per 100K Population'], s=50, alpha=0.7, color = 'green')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.ylabel('Death (per 100K Population)', fontsize=20)
plt.xlabel('Infection (per 100K Population)', fontsize=20)
plt.show()
```

```
sydata = pd.read_csv("/covid19-project/final file/data/sycluster.csv")
sydata['Positives Case per 100K Population'] = sydata['Positives Case per 10K Population'] * 10
sydata['Deaths per 100K Population'] = sydata['Deaths per 10K Population'] * 10

sydatamod = sydata.sort_values('Deaths per 10K Positive Case', ascending=False)
sydatamod['Color'] = sydatamod['PosDeathCluster'].map(lambda x: 'red' if (x == 1) else ('orange' if (x == 2) else 'green'))

plt.figure(figsize = (12,24))
plt.barh(sydatamod['County'], sydatamod['Deaths per 10K Positive Case'], height=0.8, color=sydatamod['Color'])
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.ylabel('County', fontsize=20)
plt.xlabel('Death (per 10K Infections)', fontsize=20)
plt.margins(y=0.01)
plt.show()
```
