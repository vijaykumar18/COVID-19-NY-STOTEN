```
import pandas as pd
import matplotlib.pyplot as plt
```

### Figure 1

#### Figure 1A

```
sydata = pd.read_csv("/covid19-project/final file/data/sycluster.csv")
sydata['Positives Case per 100K Population'] = sydata['Positives Case per 10K Population'] * 10
sydata['Deaths per 100K Population'] = sydata['Deaths per 10K Population'] * 10
sydatamod = sydata.sort_values('Positives Case per 100K Population', ascending=False)

plt.figure(figsize = (12,24))
plt.barh(sydatamod[sydatamod['CaseCluster'] == 1 ]['County'], sydatamod[sydatamod['CaseCluster'] == 1 ]['Positives Case per 100K Population'], height=0.8, color='red', label = 'Cluster 1')
plt.barh(sydatamod[sydatamod['CaseCluster'] == 2 ]['County'], sydatamod[sydatamod['CaseCluster'] == 2 ]['Positives Case per 100K Population'], height=0.8, color='orange', label = 'Cluster 2')
plt.barh(sydatamod[sydatamod['CaseCluster'] == 3 ]['County'], sydatamod[sydatamod['CaseCluster'] == 3 ]['Positives Case per 100K Population'], height=0.8, color='green', label = 'Cluster 3')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.ylabel('County', fontsize=20)
plt.xlabel('Infection (per 100K Population)', fontsize=20)
plt.margins(y=0.01)
plt.show()
```
